#Call Data

```python
# Load the dataset
df_store = pd.read_csv('/content/drive/MyDrive/RecSystem_Steam/3/steam.csv')  # Replace with your file path
print("Dataset Loaded:")
print(df_store.info())
print(df_store.head())
```

#Handle Missing Values

```python
# Handle missing values
df_store['categories'].fillna('', inplace=True)
df_store['genres'].fillna('', inplace=True)
df_store['platforms'].fillna('', inplace=True)

# Check for any remaining missing values
print("Missing Values Handled:")
print(df_store.isnull().sum())
```

#One-Hot Encode the categories, genres, and platforms Columns

```python
# Function to split semicolon-separated values and one-hot encode
def split_and_one_hot_encode(column, prefix):
    # Split the column and create a binary matrix
    binary_matrix = df_store[column].str.get_dummies(sep=';')
    # Rename columns with a prefix
    binary_matrix.columns = [f"{prefix}_{col}" for col in binary_matrix.columns]
    return binary_matrix

# One-hot encode 'categories', 'genres', and 'platforms'
categories_encoded = split_and_one_hot_encode('categories', 'categories')
genres_encoded = split_and_one_hot_encode('genres', 'genres')
platforms_encoded = split_and_one_hot_encode('platforms', 'platforms')

# Combine the encoded columns with the original dataset
df_store = pd.concat([df_store, categories_encoded, genres_encoded, platforms_encoded], axis=1)

print("One-Hot Encoded Features Added:")
print(df_store.head())
```

#Feature Weighting

```python
# Apply new weights to features
weighted_features = df_store.loc[:, df_store.columns.str.startswith(('categories_', 'genres_', 'platforms_'))].copy()

# Reassess and update weights
weighted_features.loc[:, weighted_features.columns.str.startswith('categories_')] *= 2  # Increase weight for categories
weighted_features.loc[:, weighted_features.columns.str.startswith('genres_')] *= 1.5  # Moderate weight for genres
weighted_features.loc[:, weighted_features.columns.str.startswith('platforms_')] *= 1  # Default weight for platforms

print("Updated Feature Weights Applied.")
print(weighted_features.head())
```

##Normlization

```python
# Normalize numerical features for inclusion
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Normalize numerical features and add them to weighted features
weighted_features['positive_ratings'] = scaler.fit_transform(df_store[['positive_ratings']])
weighted_features['price'] = scaler.fit_transform(df_store[['price']])
weighted_features['average_playtime'] = scaler.fit_transform(df_store[['average_playtime']])

print("Additional Features (popularity, price, playtime) Added.")
```

##Game-to-Game Recommendations

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Compute cosine similarity between all games
game_similarity = cosine_similarity(weighted_features)

# Function to recommend games similar to a given game
def recommend_similar_games(game_index, num_recommendations=5):
    # Get similarity scores for the selected game
    similarity_scores = list(enumerate(game_similarity[game_index]))
    # Sort games by similarity score (highest first)
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    # Get the top 'num_recommendations' (excluding the game itself)
    top_games = similarity_scores[1:num_recommendations+1]
    # Return the names and similarity scores of the recommended games
    recommended_games = [(df_store.iloc[i[0]]['name'], i[1]) for i in top_games]
    return recommended_games
```

```python
# Example: Recommend games similar to the first game in the dataset
game_index = 0  # Index of the game to base recommendations on
game_recommendations = recommend_similar_games(game_index=game_index, num_recommendations=5)

print(f"Recommendations for '{df_store.iloc[game_index]['name']}':")
for game, score in game_recommendations:
    print(f"  - {game} (Similarity: {score:.2f})")
```

```python
# Function to dynamically generate relevant games
def get_dynamic_relevant_games(user_liked_games_indices, num_relevant=5):
    """
    Generate a dynamic list of relevant games based on user preferences.
    """
    relevant_games = set()
    for game_index in user_liked_games_indices:
        # Get recommendations for each liked game
        recommendations = recommend_similar_games(game_index, num_recommendations=num_relevant)
        # Add recommended game indices to the relevant set
        relevant_games.update([df_store[df_store['name'] == game[0]].index[0] for game in recommendations])
    return list(relevant_games)
```

#Create a User Profile

```python
from sklearn.metrics.pairwise import cosine_similarity

# Simulate user preferences (e.g., indices of games the user likes)
user_liked_games_indices = [5, 12, 20]  # Indices of strategy games (or change based on your scenario)
user_liked_games = weighted_features.iloc[user_liked_games_indices]

# Create the user profile
user_profile = user_liked_games.mean(axis=0).values.reshape(1, -1)
print("User Profile Created!")
print(user_profile)

# Automatically generate relevant games
relevant_games = get_dynamic_relevant_games(user_liked_games_indices, num_relevant=5)
print(f"Relevant Games Automatically Generated: {relevant_games}")
```

#Add Filters

```python
# Update the filtering function
def recommend_with_updated_filters(user_profile, num_recommendations=5, price_range=(0, 20), platform='windows', max_age=18):
    filtered_games = df_store[
        (df_store['price'] >= price_range[0]) &
        (df_store['price'] <= price_range[1]) &
        (df_store[f'platforms_{platform}'] == 1) &
        (df_store['required_age'] <= max_age)
    ]
    filtered_indices = filtered_games.index.tolist()
    filtered_similarity_scores = cosine_similarity(weighted_features.iloc[filtered_indices], user_profile).flatten()
    top_indices = np.argsort(filtered_similarity_scores)[-num_recommendations:][::-1]
    recommendations = [(df_store.iloc[filtered_indices[i]]['name'], filtered_similarity_scores[i]) for i in top_indices]
    return recommendations
```

```python
# Example: Filtered recommendations
filtered_recommendations = recommend_with_updated_filters(
    user_profile=user_profile,
    num_recommendations=5,
    price_range=(0, 10),
    platform='windows',
    max_age=18
)

print("Filtered Recommendations Updated:")
for game, score in filtered_recommendations:
    print(f"  - {game} (Similarity: {score:.2f})")
```

#Evaluate

```python
# Function to evaluate recommendations dynamically based on user preferences
def evaluate_recommendations(recommendations, user_liked_games_indices, k, num_relevant):
    # Dynamically generate relevant games based on user liked indices
    relevant_games = [idx for idx in user_liked_games_indices[:num_relevant]]

    # Extract recommended game indices from recommendations
    recommended_indices = [df_store[df_store['name'] == game[0]].index[0] for game in recommendations[:k]]

    # Calculate relevant recommendations
    relevant_recommendations = len(set(recommended_indices) & set(relevant_games))

    # Precision@K
    precision = relevant_recommendations / k if k > 0 else 0

    # Recall@K
    recall = relevant_recommendations / len(relevant_games) if len(relevant_games) > 0 else 0

    return precision, recall, relevant_games
```

```python
precision, recall, relevant_games = evaluate_recommendations(
    filtered_recommendations, user_liked_games_indices, k, num_relevant
)
print(f"Dynamically Generated Relevant Games: {relevant_games}")
print(f"Precision@{k}: {precision:.2f}")
print(f"Recall@{k}: {recall:.2f}")
```

```python
import gradio as gr
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load preprocessed dataset and features
df_store = pd.read_csv('/content/drive/MyDrive/RecSystem_Steam/preprocessed_dataset_final.csv')
weighted_features = pd.read_csv('/content/drive/MyDrive/RecSystem_Steam/weighted_features_final.csv')

# Compute cosine similarity matrix for games
game_similarity = cosine_similarity(weighted_features)

# Recommendation functions
def recommend_games(profile_name, num_recommendations):
    # Define user profiles
    user_liked_games_indices = {
        "FPS Enthusiast": [0, 3, 8],   # Replace with appropriate indices
        "Strategy Lover": [2, 10, 12],
        "Budget Gamer": [5, 12, 15],
    }

    # Get indices of liked games based on the profile
    indices = user_liked_games_indices.get(profile_name, [])
    user_liked_games = weighted_features.iloc[indices]

    # Create a user profile
    user_profile = user_liked_games.mean(axis=0).values.reshape(1, -1)

    # Compute similarity scores
    user_similarity_scores = cosine_similarity(weighted_features, user_profile).flatten()

    # Get top recommendations
    recommended_indices = user_similarity_scores.argsort()[-num_recommendations:][::-1]
    recommendations = [(df_store.iloc[i]['name'], round(user_similarity_scores[i], 2)) for i in recommended_indices]
    return recommendations

def recommend_similar_games(game_name, num_recommendations):
    # Ensure the game exists in the dataset
    if game_name not in df_store['name'].values:
        return [("Game not found", 0)]

    # Get the index of the selected game
    game_index = df_store[df_store['name'] == game_name].index[0]

    # Compute similarity scores for the selected game
    similarity_scores = list(enumerate(game_similarity[game_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Get top recommendations excluding the game itself
    top_games = similarity_scores[1:num_recommendations+1]
    recommendations = [(df_store.iloc[i[0]]['name'], round(i[1], 2)) for i in top_games]
    return recommendations

# Combined function for Gradio interface
def combined_recommendation(mode, profile_name, game_name, num_recommendations):
    if mode == "User Profile Recommendation":
        recommendations = recommend_games(profile_name, num_recommendations)
    elif mode == "Game-to-Game Recommendation":
        recommendations = recommend_similar_games(game_name, num_recommendations)
    else:
        return "Invalid mode selected."

    # Format the results
    result = "\n".join([f"{game} (Similarity: {score})" for game, score in recommendations])
    return result

# Define Gradio interface
modes = ["User Profile Recommendation", "Game-to-Game Recommendation"]
profiles = ["FPS Enthusiast", "Strategy Lover", "Budget Gamer"]
game_list = list(df_store['name'])  # List of games in the dataset

interface = gr.Interface(
    fn=combined_recommendation,
    inputs=[
        gr.Dropdown(choices=modes, label="Select Recommendation Mode"),
        gr.Dropdown(choices=profiles, label="Select User Profile"),
        gr.Dropdown(choices=game_list, label="Select Game"),
        gr.Slider(minimum=1, maximum=10, step=1, value=5, label="Number of Recommendations"),
    ],
    outputs="text",
    title="Game Recommendation System",
    description="Choose a recommendation mode: User Profile or Game-to-Game. Get personalized recommendations!"
)
```
